```python
df_counted = pd.DataFrame(df_test.groupby(['rd_created_at','CVE'])['CVE'].count())
print(df_counted)
```

```python
print(type(df_counted))
```

```python
df_counted.rename(columns = {'CVE':'count'}, inplace = True)
```

```python
df_counted
```

```python
df_counted.reset_index(level = -1, inplace = True)
df_counted
```

```python
df_48hour = pd.DataFrame()
```

```python
import datetime
import numpy as np
for i, cve in enumerate(CVE_list):
    
    # create null DF
    df_48hour_ = pd.DataFrame()
    
    # Define the start & end time (48hours) for the cve-id
    start_time = df_counted[df_counted['CVE'] == cve].index[0]
    end_time = start_time + datetime.timedelta(hours = 48)

    # Create a temporary Dataframe for the cve
    df_48hour_ = df_counted[df_counted['CVE'] == cve].loc[start_time:end_time]

    # Concatenate temporary Dataframes
    if (i == 0):
        df_48hour = df_48hour_
        
    else:
        df_48hour = pd.concat([df_48hour, df_48hour_])
df_48hour
```

```python
df_ = df_48hour.reset_index()
df_
```
