```python
# Read the csv files
import pandas as pd
import datetime
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
```

```python
df_features = pd.read_csv("1677CVEs_783dim_features_20230326.csv")
df_features
```

```python
df_flag = pd.read_csv("CVE_list_with_flagH_20230326.csv", index_col = ['CVE'] )
df_flag_sorted = df_flag.sort_values('CVE')
df_flag_sorted
```

```python
df_flag_sorted.drop('Unnamed: 0', inplace = True, axis = 1)
```

```python
df_flag_sorted
```

```python
df_features.rename({"Unnamed: 0" : "CVE"}, inplace = True, axis = 1)
df_features
```

```python
df_features.set_index('CVE', inplace = True)
df_features
```

## Dividing Data

```python
df_features.isnull().sum()
```

## Filter the features which are NaN

```python
from tsfresh import select_features
from tsfresh.utilities.dataframe_functions import impute
impute(df_features)
y = df_flag_sorted['flag_H']
features_filterd = select_features(df_features, y)
features_filterd
```

## Random Forest (No Cross Validation)

```python
train_data = df_features
y = df_flag_sorted['flag_H']
X = train_data.values

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 1234)
```

```python
clf = RandomForestClassifier(random_state = 1234)
clf.fit(X_train, y_train)
print(f"accuracy={clf.score(X_test, y_test)}")
```

```python
y_predict = clf.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
print(f"comfusion_matrix={confusion_matrix(y_test, y_predict)}")
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
print(f"accuracy_score={accuracy_score(y_test, y_predict)}\n\
precision_score={precision_score(y_test, y_predict)}\n\
recall_score={recall_score(y_test, y_predict)}\n\
f1_score={f1_score(y_test, y_predict)}")
```
